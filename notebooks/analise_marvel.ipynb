```python
import pandas as pd

df = pd.read_csv("../data/marvel_movies.csv")
```

```python
fase_receita = df.groupby("Fase")["Bilheteria"].sum()

print(fase_receita)
```

```python
import matplotlib.pyplot as plt

fase_receita.plot(kind="bar")

plt.title("Bilheteria por Fase do MCU")
plt.xlabel("Fase")
plt.ylabel("Bilheteria (milhões USD)")
plt.xticks(rotation=0)

plt.savefig("../images/bilheteria_por_fase.png", bbox_inches="tight")

plt.show()
```

```python
top10 = df.sort_values("Bilheteria", ascending=False).head(10)

print(top10[["Filme", "Bilheteria"]])
```

```python
import matplotlib.pyplot as plt

top10 = df.sort_values("Bilheteria", ascending=False).head(10)

plt.figure(figsize=(10,6))

plt.barh(top10["Filme"], top10["Bilheteria"])

plt.title("Top 10 Filmes MCU por Bilheteria")
plt.xlabel("Bilheteria (milhões USD)")
plt.ylabel("Filme")

plt.gca().invert_yaxis()

plt.savefig("../images/top10_bilheteria.png", bbox_inches="tight")

plt.show()
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

plt.scatter(df["NotaIMDb"], df["Bilheteria"])

plt.title("IMDb x Bilheteria")
plt.xlabel("Nota IMDb")
plt.ylabel("Bilheteria (milhões USD)")

plt.savefig("../images/imdb_vs_bilheteria.png", bbox_inches="tight")

plt.show()
```

```python
correlacao = df["NotaIMDb"].corr(df["Bilheteria"])

print("Correlação:", round(correlacao, 2))
```

```python
df["Eficiencia"] = df["Bilheteria"] / df["NotaIMDb"]

top_eficiencia = df.sort_values(
    "Eficiencia",
    ascending=False
)

print(
    top_eficiencia[
        ["Filme", "NotaIMDb", "Bilheteria", "Eficiencia"]
    ].head(10)
)
```
